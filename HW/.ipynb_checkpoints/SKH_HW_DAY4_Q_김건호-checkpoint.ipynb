```python
import numpy as np
import pandas as pd
```

 - sample3.csv 파일 읽고 상위 3개 데이터 조회

```python
# Code here
df=pd.read_csv('sample3.csv')
df.head(3)
```

 - df의 columns 조회

```python
# Code here
df.columns
```

 - 열 이름에 붙어 있는 탭 (\t) 문자 제거
   - 문자열 처리 방법 series/df.columns/df.index.str.함수

```python
# Code here
df.columns = df.columns.str.strip()
df.columns
```

 - X4, X20 열의 데이터에 붙어 있는 탭 (\t) 문자 제거 후 상위 3개 출력

```python
# Code here
df['X4'] = df['X4'].str.strip()
df['X20'] = df['X20'].str.strip()
# df[['X4','X20']]=df[['X4','X20']].str.strip('\t') #이렇게 한번에 하는건 오류
# df[['X4','X20']]=df[['X4','X20']].str.replace('\t','') #이렇게 한번에 하는건 오류
df.head(3)
```

 - df의 정보 조회

```python
# Code here
df.info()
```

 - X4 열의 데이터 타입을 float64로 변환
   - 변환 과정에서 발생하는 오류를 해결하시오 (to_numeric 등)
   - 변환 후 발생하는 결측치는 0으로 채우시오.

```python
# Code here
df['X4'] = pd.to_numeric(df['X4'])
df['X4'] = df['X4'].fillna(0)
df['X4']
```

 - X4 열의 데이터 타입 조회

```python
# Code here
df['X4'].dtype
```

 - df의 통계 요약 조회

```python
# Code here
df.describe()
```

 - 일부 열 이름 변경
   - Y: refresh, X1: Wafer_Site, X2: ISO_Etch_Depth, X3: BG_Etch_Depth, X4: BG_Etch_Height, X5: Ohm, X20: Zone

```python
# Code here
df = df.rename(columns={'Y': 'refresh', 'X1': 'Wafer_Site', 'X2': 'ISO_Etch_Depth', 'X3': 'BG_Etch_Depth', 
                        'X4': 'BG_Etch_Height', 'X5': 'Ohm', 'X20': 'Zone'})
df
```

 - refresh 열을 기준으로 내림차순 정렬 후 상위 3개 출력 (Top 3)

```python
# Code here
df = df.sort_values(by='refresh', ascending=False)
df.head(3)
```

 - refresh 열의 통계 요약 조회

```python
# Code here
df['refresh'].describe()
```

 - Wafer_site 열의 고윳값 조회

```python
# Code here
df['Wafer_Site'].unique()
```

 - Wafer_site 열의 고윳값 별 데이터 수 조회

```python
# Code here
df['Wafer_Site'].value_counts()
```

 - Wafer_site의 값이 5인 행 추출하여 df2에 저장 후 출력

```python
# Code here
#df2 = list(np.where(df['Wafer_Site']==5)).index.tolist()

df2 = df[df['Wafer_Site']==5]
df2
```

 - df2의 refresh 열의 최솟값, 최댓값, 평균, 중앙값 소수점 셋째자리까지 출력

```python
df2['refresh'].apply('min').round(3)
```

```python
# Code here
print(f"min: {round(df2['refresh'].min(),3)}")
print(f"max: {round(df2['refresh'].max(),3)}")
print(f"mean: {round(df2['refresh'].mean(),3)}")
print(f"median: {round(df2['refresh'].median(),3)}")
```

 - df2의 데이터 중 refresh 값이 260 이상인 데이터를 refresh 값을 기준으로 내림차순 정렬

```python
# Code here
df2 = df2[df2['refresh']>=260].sort_values(by='refresh', ascending=False)
df2
```
